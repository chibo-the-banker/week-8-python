```python
world = gpd.read_file(geodatasets.get_path(".../python3.8/site-packages/geopandas/datasets/naturalearth_lowres/naturalearth_lowres.shp"))
```

```python
world.head(10)
```

```python
print("Column list: ", world.columns.tolist())
```

```python
world.plot()
```

```python
world = world.dissolve(by='CONTINENT').reset_index()

# Merge with your data
merged = world.merge(countriesMap, 
                           left_on='CONTINENT', 
                           right_on='continent', 
                           how='left')

# Plot
fig, ax = plt.subplots(figsize=(15, 10))
merged.plot(column='total_deaths',
           ax=ax,
           legend=True,
           legend_kwds={'label': "Your Metric", 'orientation': "horizontal"},
           cmap='OrRd',
           edgecolor='black',
           linewidth=0.5)

ax.set_title('Your Metric by Continent', fontsize=16)
ax.set_axis_off()  # Remove axes for cleaner look
plt.tight_layout()
plt.show()
```
